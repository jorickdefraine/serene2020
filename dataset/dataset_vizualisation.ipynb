```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#import cv2
```

```python
train_file = open("train.bin","rb")
test_file = open("test.bin","rb")

train_num = list(train_file.read())
test_num = list(test_file.read())

train_labels = np.array(train_num[0::617])
test_labels = np.array(test_num[0::617])
```

```python
cpt = 0
for j in range(20):
    print (cpt)
    for i in range(len(train_labels)):
        if train_labels[i] == j:
            cpt+=1
print(cpt)
    
```

```python
train_hist = np.histogram(train_labels, bins=np.linspace(-1,20,21))
test_hist = np.histogram(test_labels, bins=np.linspace(-1,20,21))
```

```python
plt.hist(train_labels, bins=np.linspace(-1,20,21))
plt.title("Distribution of the labels in train dataset")
plt.show()
print("sum : ",sum(train_hist[0]))
print(train_hist[0])
plt.close()
```

```python
plt.hist(test_labels, bins=np.linspace(-1,20,21))
plt.title("Distribution of the labels in test dataset")
plt.show()
print("sum : ",sum(test_hist[0]))
print(test_hist[0])
plt.close()
```
